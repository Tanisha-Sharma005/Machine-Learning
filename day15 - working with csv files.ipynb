# Index_col parameter

```python
pd.read_csv(aug_train,index_col='enrollee_id')
```

# Header Parameter

```python
test="https://people.sc.fsu.edu/~jburkardt/data/csv/airtravel.csv"
pd.read_csv(test,header=1)
```

# use_cols parameter

```python
pd.read_csv(aug_train,usecols=['enrollee_id','gender','education_level'])
```

# Squeeze parameter

```python
#pd.read_csv(aug_train,usecols=['gender'],Squeeze=True)
pd.read_csv(aug_train,usecols=['gender'])['gender']
```

# Skiprows/nrows parameter

```python
pd.read_csv(aug_train,skiprows=[0,53])
pd.read_csv(aug_train,nrows=100)
```

# Encoding Parameter

```python
zomato = "https://raw.githubusercontent.com/AbhishekNagaraj1/Zomato-Data-Analysis/master/zomato.csv"
pd.read_csv(zomato,encoding='latin-1')
```

# skip bad lines

```python
BX_Books = "https://raw.githubusercontent.com/thehawkgriffith/book-recommender/master/BX-Books.csv"
pd.read_csv(BX_Books,sep=';',encoding='latin-1',on_bad_lines='skip')
```

# dtypes of parameter

```python
pd.read_csv(aug_train).info()
pd.read_csv(aug_train,dtype={'target':int})
```
